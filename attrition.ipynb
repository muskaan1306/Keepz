```python
import pandas as pd
import random
from datetime import datetime, timedelta
```

```python

NUM_EMPLOYEES = 1000

employees = []

departments = [
    'Sales', 'Marketing', 'Finance',
    'Operations', 'HR', 'Engineering', 'Support'
]

job_levels = ['Junior', 'Mid-Level', 'Senior', 'Lead']
genders = ['Male', 'Female']

today = datetime.today()

for i in range(NUM_EMPLOYEES):

    employee = {}

    employee['emp_id'] = f"E_{i+1:05d}"

    employee['gender'] = random.choice(genders)
    employee['age'] = random.randint(22, 60)
    employee['department'] = random.choice(departments)
    employee['job_level'] = random.choice(job_levels)
    employee['years_of_service'] = random.randint(0, 20)
    employee['years_in_current_role'] = random.randint(0,employee['years_of_service'])
    employee['time_since_joining_team'] = random.randint(0, 8)

    employee['overtime_frequency'] = random.randint(0, 20)
    employee['average_weekly_hours'] = random.randint(35, 70)
    employee['after_hours_work'] = random.randint(0, 15)
    employee['weekend_work'] = random.randint(0, 8)
    employee['active_projects'] = random.randint(1, 10)
    employee['recent_workload_increase_pct'] = random.randint(-10, 60)

    employee['engagement_score'] = round(random.uniform(1, 5), 1)
    employee['pulse_survey_score'] = round(random.uniform(1, 5), 1)
    employee['job_satisfaction'] = round(random.uniform(1, 5), 1)
    employee['manager_feedback_sentiment'] = round(random.uniform(1, 5),1)
    employee['team_participation_score'] = random.randint(1, 100)
    employee['engagement_trend'] = round(random.uniform(-2, 2),2)

    employee['months_since_last_promotion'] = random.randint(0, 72)
    employee['internal_job_applications'] = random.randint(0, 5)
    employee['training_participation'] = random.randint(0, 10)
    employee['career_path_progress_pct'] = random.randint(0, 100)
    employee['certifications_completed'] = random.randint(0, 8)
    employee['has_mentor'] = random.choice([0, 1])

    employee['monthly_salary'] = random.randint(30000,350000)
    employee['salary_percentile'] = random.randint(1, 100)
    employee['months_since_last_salary_increase'] = random.randint(0,36)
    employee['bonus_history_score'] = random.randint(1, 5)
    employee['pay_competitiveness_score'] = random.randint(1, 100)
    employee['internal_salary_gap_pct'] = round(random.uniform(-30, 30),2)
    
    employee['unplanned_leaves'] = random.randint(0, 15)
    employee['sick_leave_days'] = random.randint(0, 20)
    employee['leave_pattern_change_pct'] = round(random.uniform(-50, 100),2)

    employee['peer_recognition_count'] = random.randint(0, 20)
    employee['awards_received'] = random.randint(0, 5)
    employee['manager_appreciation_frequency'] = random.randint(0, 12)

    employee['manager_changes_last_2yr'] = random.randint(0, 4)
    employee['manager_quality_score'] = round(random.uniform(1, 5),1)
    employee['team_attrition_rate_pct'] = round(random.uniform(0, 35),1)

    employee['commute_time_minutes'] = random.randint(5, 120)
    employee['work_life_balance_score'] = round(random.uniform(1, 5),1)
    employee['work_flexibility_score'] = random.randint(1, 5)


    employee['skill_criticality_score'] = random.randint(1, 5)
    employee['replacement_cost'] = random.randint(50000,1500000)
    employee['ramp_up_months'] = random.randint(1,12)
    employee['knowledge_risk_score'] = random.randint(1, 5)
    employee['succession_coverage'] = random.choice(
        [
            'Ready Successor',
            'Partial Backup',
            'No Backup'
        ]
    )
    employee['annual_revenue_contribution'] = random.randint(0,50000000)

    employee['performance_rating'] = random.randint(1, 5)
    employee['high_performer_flag'] = (1 if employee['performance_rating'] >= 4 else 0)

    joining_days_ago = random.randint(0, 3650)
    employee['joining_date'] = (today - timedelta(days=joining_days_ago)).date()
    employees.append(employee)
```

```python
df = pd.DataFrame(employees)

df.to_csv("employee_attrition_dataset.csv", index=False)

df.head()
```
